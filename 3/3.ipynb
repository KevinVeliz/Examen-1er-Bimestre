```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
import pandas as pd
import bson
from bson.raw_bson import RawBSONDocument
```

```python
db_client = MongoClient()
my_db = db_client.cursos
my_posts = my_db.posts
    
def find_2nd(string, substring):
    return string.find(substring, string.find(substring) + 1)
def find_1st(string, substring):
    return string.find(substring, string.find(substring))
```

```python
response = requests.get("https://olympics.com/tokyo-2020/es/noticias/videos/mariana-pajon-medalla-de-plata-para-colombia-en-ciclismo-bmx-ramirez-con-bronce-")
soup = BeautifulSoup(response.content, "lxml")

title=[]
paragraph=[]

post_title = soup.find_all("h1", class_="tk-article__title")
post_prh = soup.find_all("a", class_="tk-nav-footer__link")

#extracted = []
    
for element in post_title:
    #print(element)
    element=str(element)
    limpio=str(element[find_1st(element, '>')+1:find_2nd(element,'<')])
    #print (limpio)
    title.append(limpio.strip())
    
for element in post_prh:
    #print(element)
    element=str(element)
    limpio=str(element[find_1st(element, '>')+1:find_2nd(element,'<')])
    #print (limpio)
    paragraph.append(limpio.strip())

print(title)
print(post_title)


print(paragraph)
print(post_prh)
```
